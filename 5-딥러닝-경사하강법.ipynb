```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 공부시간 X와 성적 Y 데이터
#data = [[1, 10], [3, 30], [5, 50], [7, 70]]
data = [[2, 81], [4, 93], [6, 91], [8, 97]]

x = np.array([i[0] for i in data])
y = np.array([i[1] for i in data])
x, y
```

```python
# 산점도 그리기
plt.scatter(x, y)
plt.show()
```

```python
lr = 0.01       # 학습률
W, b = 1, 0     # 가중치와 절편 초기화
epochs = 1000    # 에포크
```

```python
# 경사 하강법 구현
for i in range(epochs):
  y_hat = W * x + b    # 예측값
  #error = y_hat - y    # 오차
  error = y - y_hat
  loss = (error ** 2).mean()  # 손실값

  # 손실함수(MSE)를 W, b로 미분한 값 (그레디언트)
  #W_grad = -(2 / n) * sum(x * error)
  #b_grad = -(2 / n) * sum(error)
  W_grad = (error * x).mean()
  b_grad = error.mean()
  
  W = W + lr * W_grad
  b = b + lr * b_grad

  if i % 100 == 0:
    print(f'epoch={i}, 기울기={W}, 절편={b}, loss={loss}')
```

```python
# 그래프 다시 그리기
plt.scatter(x, y)
plt.plot([1, 7], [1 * W + b, 7 * W + b])
plt.show()
```

```python
# 회귀식 직접 구하기 (학습지3 - 회귀분석에서 도출했던 식)
mean_x = np.mean(x)
mean_y = np.mean(y)

# 기울기 찾기
bottom = np.sum(((x - mean_x) ** 2))
top = np.sum((x - mean_x) * (y - mean_y))
W2 = top / bottom

# 절편 찾기
b2 = mean_y - W2 * mean_x

print(W2, b2)   # 경사하강법에서 나온 값과 확인하기
```
