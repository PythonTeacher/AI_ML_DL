```python
# AND gate
def AND(x1, x2):    
    w1, w2 = 0.6, 0.6  # 입력값 x1에 대한 가중치 w1, x2에 대한 가중치 w2    
    b = -0.6          # bias 편향값
    
    result = x1 * w1 + x2 * w2 + b  # 연산 결과

    # 결과가 0보다 작거나 같으면 0, 0보다 크면 1을 반환합니다.
    return 0 if result <= 0 else 1
```

```python
input_list = [(0, 0), (0, 1), (1, 0), (1, 1)]

for x1, x2 in input_list:
    result = AND(x1, x2)
    print(f'입력값: {x1} {x2} / AND 결과: {result}')
```

```python
# OR gate
def OR(x1, x2):
    w1, w2 = 0.6, 0.6  # 입력값 x1에 대한 가중치 w1, x2에 대한 가중치 w2    
    b = 0           # bias 편향값    

    result = x1 * w1 + x2 * w2 + b   # 연산 결과

    # 결과가 0보다 작거나 같으면 0, 0보다 크면 1을 반환합니다.
    return 0 if result <= 0 else 1
```

```python
for x1, x2 in input_list:
    result = OR(x1, x2)
    print(f'입력값: {x1} {x2} / OR 결과: {result}') 
```

```python
# NAND gate
def NAND(x1, x2):
    return 1 if AND(x1, x2) == 0 else 0 
```

```python
for x1, x2 in input_list:
    result = NAND(x1, x2)    
    print(f'입력값: {x1} {x2} / NAND 결과: {result}')
```

```python
# XOR Gate
for x1, x2 in input_list:    
    or_result = OR(x1, x2)
    nand_result = NAND(x1, x2)
    xor_result = AND(or_result, nand_result)

    print(f'입력값: {x1} {x2} / XOR 결과: {xor_result}')
```

```python
def XOR(x1, x2):
  result = AND(OR(x1, x2), NAND(x1, x2))
  return result
```

```python
for x1, x2 in input_list:
  result = XOR(x1, x2)
  print(f'입력값: {x1} {x2} / XOR 결과: {result}')
```

```python
import numpy as np

def init_network():
  network = {}
  network['W1'] = np.array([[0.1, 0.3, 0.5], [0.2, 0.4, 0.6]])
  network['b1'] = np.array([0.1, 0.2, 0.3])
  network['W2'] = np.array([[0.1, 0.2, 0.3]])
  network['b2'] = np.array([0.1])

  return network
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def forward(network, x):
  W1, W2 = network['W1'], network['W2']
  b1, b2 = network['b1'], network['b2']

  a1 = np.dot(x, W1) + b1
  z1 = sigmoid(a1)
  #a2 = np.dot(z1, W2.T) + b2
  a2 = np.dot(z1, W2.reshape(3, 1)) + b2
  y = sigmoid(a2)

  return y
```

```python
network = init_network()
x = np.array([1.0, 0.5])
y = forward(network, x)
print(y)
```
