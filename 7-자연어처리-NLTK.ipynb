## 2. 토큰 생성 (Tokenizing)

자연어 처리를 위해서는 긴 문자열을 작은 단위로 나누어야 한다.

이 문자열 단위를 토큰(token)이라고 하고, 토큰으로 나누는 작업을 토큰 생성(tokenizing)이라고 한다.

단어나 문장을 토큰으로 사용하거나 다양한 Tokenizer를 사용하여 토큰을 생성할 수 있다.

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
text = """
She was the youngest of the two daughters of a most affectionate,
indulgent father; and had, in consequence of her sister's marriage,
been mistress of his house from a very early period.  Her mother
had died too long ago for her to have more than an indistinct
remembrance of her caresses; and her place had been supplied
by an excellent woman as governess, who had fallen little short
of a mother in affection.
"""
word_tokens = word_tokenize(text) 
print(word_tokens)
```

```python
from nltk.tokenize import RegexpTokenizer
retokenize = RegexpTokenizer("[\w]+")
print(retokenize.tokenize(text))
```

```python
sent_tokens = sent_tokenize(text)
print(sent_tokens)
```

```python
text = "NLTK는 Natural Language Toolkit의 줄임말입니다 Python 프로그래밍 언어로 작성된 영어의 기호 및 통계 자연 언어 처리를위한 라이브러리 및 프로그램 모음입니다."

word_tokens = word_tokenize(text)
sent_tokens = sent_tokenize(text)

print(word_tokens)
print(sent_tokens)
```

```python
text = "I am actively looking for Ph.D. students. and you are a Ph.D student."
print(sent_tokenize(text))
```

## 3. 형태소 분석 (Morphological Analysis)

형태소는 언어학에서 일정한 의미가 있는 가장 작은 말의 단위를 뜻한다.

형태소 분석(morphological analysis)이란 단어로부터 어근, 접두사, 접미사, 품사 등 다양한 언어적 속성을 파악하고 이를 이용하여 형태소를 찾아내거나 처리한다.

* 어간 추출 (Stemming)
* 원형 복원 (Lemmatizing)

```python
from nltk.stem import PorterStemmer, LancasterStemmer
```

```python
st1 = PorterStemmer()
st2 = LancasterStemmer()

words = ['fly', 'flies', 'flying', 'flew', 'flown']

print([st1.stem(w) for w in words])
print([st2.stem(w) for w in words])
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('omw-1.4')
```

```python
lm = WordNetLemmatizer()

print([lm.lemmatize(w, pos="v") for w in words])    # pos(part of speech) : 품사 지정
```

## 4. 품사 태깅 (Part-of-speech(POS) Tagging)

품사는 낱말을 문법적인 기능이나 형태, 뜻에 따라 구분한 것이다.

NLTK에서는 Penn Treebank Tagset을 사용하여 아래의 품사로 구분한다.

* NNP : 단수 고유명사
* VB : 동사
* VBP : 동사 현재형
* TO : to 전치사
* NN : 명사(단수 or 복수)
* DT : 관형사

```python
from nltk.tag import pos_tag
```

```python
text = "Emma refused to permit us to obtain the refuse permit"
tagged_list = pos_tag(word_tokenize(text))
tagged_list
```

```python
nouns_list = [t[0] for t in tagged_list if t[1] == "NN"]
nouns_list
```

# 기타 클래스

* Text 클래스 : 문서 분석에 유용한 기능 제공
* WordCloud 클래스 : 사용 빈도수에 따라 워드클라우드 시각화

```python
from nltk import Text
import matplotlib.pyplot as plt
```

```python
# 상위 20개 단어 보기
text = Text(retokenize.tokenize(emma_raw_data))

plt.figure(figsize=(10, 5))
text.plot(20)
plt.show()
```

```python
# 단어가 사용된 위치 표시
text.concordance("Emma")
```

```python
from collections import Counter

# 불용어 정의
stopwords = ["Mr.", "Mrs.", "Miss", "Mr", "Mrs", "Dear"]

# 품사 태깅
emma_tokens = pos_tag(retokenize.tokenize(emma_raw_data))

# 고유명사이면서 stop word에 없는 단어 추출
names_list = [t[0] for t in emma_tokens if t[1] == "NNP" and t[0] not in stopwords]
counts = Counter(names_list)
```

```python
# 가장 많이 나온 이름 50개 저장
top_names = counts.most_common(50)
print(top_names)
```

```python
from wordcloud import WordCloud

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)

plt.figure(figsize=(10, 8))
plt.imshow(wc.generate_from_frequencies(dict(top_names)))    # {'Emma': 830, 'Harriet': 491, ..}
plt.axis("off")
plt.show()
```
